Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp

```python
def dem(danh_sach):
    if not danh_sach:
        return None

    cac_so_duy_nhat = []

    for so in danh_sach:
        if so not in cac_so_duy_nhat:
            cac_so_duy_nhat.append(so)

    ket_qua = []

    for so in cac_so_duy_nhat:
        so_lan = 0

        for phan_tu in danh_sach:
            if phan_tu == so:
                so_lan += 1

        ket_qua.append((so, so_lan))

    return dict(ket_qua)


def mang_dep(mang):
    tan_suat = dem(mang)

    if tan_suat is None:
        print("True")
        return

    xoa = 0
    mang_moi = []

    for so, so_lan in tan_suat.items():
        if so_lan < so:
            xoa += so_lan

        elif so_lan > so:
            xoa += (so_lan - so)
            for _ in range(so):
                mang_moi.append(so)
        else:
            for _ in range(so):
                mang_moi.append(so)

    if xoa == 0:
        print("True")
    else:
        print("False")
        print("Số lượng cần xóa:", xoa)
        print("Mảng mới:", mang_moi)


test = list(map(int, input().split()))

mang_dep(test)
```

Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có phần tử nào thỏa mãn thì không in gì

```python
def boyer_moore(mang):
    n = len(mang)
    if n == 0:
        return
    
    bien_tam = None
    count = 0
    
    for x in mang:
        if count == 0:
            bien_tam = x
            count = 1
        elif x == bien_tam:
            count += 1
        else:
            count -= 1
            
    so_lan = 0
    for x in mang:
        if x == bien_tam:
            so_lan += 1
            
    if so_lan > n // 2:
        print(bien_tam)

a = list(map(int, input().split()))
boyer_moore(a)
```

Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng dần, lẻ giảm dần

```python
def sap_xep(mang):
    chan = []
    le = []
    
    for so in mang:
        if so % 2 == 0:
            chan.append(so)
        else:
            le.append(so)
            
    chan.sort()
    le.sort(reverse=True)
    
    ket_qua = chan + le
    return ket_qua

mang_ban_dau = list(map(int, input().split()))
mang_moi = sap_xep(mang_ban_dau)
print(mang_moi)
```

Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào

```python
def merge(list1, list2):
    ket_qua = []

    i = 0
    j = 0
    tang = True

    if len(list1) >= 2:
        tang = list1[0] <= list1[1]
    elif len(list2) >= 2:
        tang = list2[0] <= list2[1]

    while i < len(list1) and j < len(list2):

        if tang:
            if list1[i] <= list2[j]:
                ket_qua.append(list1[i])
                i += 1
            else:
                ket_qua.append(list2[j])
                j += 1

        else:
            if list1[i] >= list2[j]:
                ket_qua.append(list1[i])
                i += 1
            else:
                ket_qua.append(list2[j])
                j += 1

    # Thêm phần còn lại
    while i < len(list1):
        ket_qua.append(list1[i])
        i += 1

    while j < len(list2):
        ket_qua.append(list2[j])
        j += 1

    return ket_qua


# Nhập dữ liệu
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))

print(merge(list1, list2))
```

Cho dãy N số nguyên dương, tính tổng các số có k chữ số. Dòng đầu ghi N số nguyên dương, dòng sau ghi số nguyên dương k

```python
def tinh(n):
    so = 0
    while n >= 1:
        n //= 10
        so += 1
    return so
N = list(map(int, input().split()))
k = int(input())
tong = 0
for so in N:
    if tinh(so) == k:
        tong += so
print(tong)
```

Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
def tim_ten(danh_sach):
    if not danh_sach:
        return None
    
    x = {}
    for ten in danh_sach:
        if ten in x:
            x[ten] += 1
        else:
            x[ten] = 1
            
    ket_qua = None
    hay_gap = -1
    
    for ten in x:
        so_lan = x[ten]
        if so_lan > hay_gap:
            hay_gap = so_lan
            ket_qua = ten
        elif so_lan == hay_gap:
            if ten < ket_qua:
                ket_qua = ten
                
    return ket_qua

sinh_vien = list(map(str, input("Nhập danh sách tên sinh viên: ").split(",")))
print(tim_ten(sinh_vien))
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài

=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái.

```python
def dem_so_luong_ten(xau):
    cac_tu = xau.split()
    so_luong = 0
    
    for tu in cac_tu:
        if len(tu) >= 2:
            if tu.isalpha():
                if tu[0].isupper() and tu[1:].islower():
                    so_luong += 1
                    
    return so_luong

van_ban = input("Nhập một đoạn văn: ")
print(dem_so_luong_ten(van_ban))
```

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số

```python
so_dau_vao = input("Nhập một số nguyên: ")
k = int(input("Nhập số k: "))

so_can_xoa = len(so_dau_vao) - k
ngan_xep = []

for chu_so in so_dau_vao:
    while so_can_xoa > 0 and ngan_xep and ngan_xep[-1] < chu_so:
        ngan_xep.pop()
        so_can_xoa -= 1
    ngan_xep.append(chu_so)

so_dau_ra = "".join(ngan_xep[:k])
print(so_dau_ra)
```

Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các số đó

```python
def tinh_tong_va_trung_binh(van_ban):
    danh_sach_so = []
    so_tam = ""
    
    for ky_tu in van_ban:
        if '0' <= ky_tu <= '9':
            so_tam += ky_tu
        else:
            if so_tam != "":
                danh_sach_so.append(int(so_tam))
                so_tam = ""
                
    if so_tam != "":
        danh_sach_so.append(int(so_tam))
        
    if not danh_sach_so:
        return 0, 0
        
    tong = 0
    for so in danh_sach_so:
        tong += so
        
    trung_binh = tong / len(danh_sach_so)
    return tong, trung_binh

van_ban_nhap = input("Nhập một đoạn văn: ")
tong, trung_binh = tinh_tong_va_trung_binh(van_ban_nhap)

print(f"Tổng: {tong}")
print(f"Trung bình: {trung_binh}")
```


Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không

```python
def kiem_tra_ngoat(xau):
    kho = []
    cap_ngoac = {')': '(', ']': '[', '}': '{'}
    
    for ky_tu in xau:
        if ky_tu in '([{':
            kho.append(ky_tu)
        elif ky_tu in ')]}':
            if not kho:
                return False
            gan_nhat = kho.pop()
            if gan_nhat != cap_ngoac[ky_tu]:
                return False
                
    if kho:
        return False
    return len(kho) == 0
day_ngoac = input("Nhập một chuỗi ngoặc: ")
if kiem_tra_ngoat(day_ngoac):
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```

Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
def tim_day_hop_so_dai_nhat(N):
    if N < 4:
        return []
    
    la_nguyen_to = [True] * (N + 1)
    la_nguyen_to[0] = la_nguyen_to[1] = False
    for p in range(2, int(N**0.5) + 1):
        if la_nguyen_to[p]:
            for i in range(p * p, N + 1, p):
                la_nguyen_to[i] = False
    
    day_max = []
    day_tam = []
    
    for i in range(2, N + 1):
        if not la_nguyen_to[i]:
            day_tam.append(i)
        else:
            if len(day_tam) > len(day_max):
                day_max = day_tam
            day_tam = []
            
    if len(day_tam) > len(day_max):
        day_max = day_tam
        
    return day_max

N_nhap = int(input())
ket_qua = tim_day_hop_so_dai_nhat(N_nhap)
if ket_qua:
    print(*(ket_qua))
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
def tim_day_con_tang_dai_nhat(mang):
    if not mang:
        return []

    day_max = []
    day_hien_tai = [mang[0]]

    for i in range(1, len(mang)):
        if mang[i] > mang[i-1]:
            day_hien_tai.append(mang[i])
        else:
            if len(day_hien_tai) > len(day_max):
                day_max = day_hien_tai
            day_hien_tai = [mang[i]]

    if len(day_hien_tai) > len(day_max):
        day_max = day_hien_tai

    return day_max

mang_nhap = list(map(int, input().split()))
ket_qua = tim_day_con_tang_dai_nhat(mang_nhap)
print(*(ket_qua))
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python

def tim_day_con_tang_khong_lien_tuc_dai_nhat(mang):
    if not mang:
        return []
    
    n = len(mang)
    f = [1] * n
    truoc = [-1] * n
    
    for i in range(1, n):
        for j in range(i):
            if mang[i] > mang[j] and f[i] < f[j] + 1:
                f[i] = f[j] + 1
                truoc[i] = j
                
    max_f = max(f)
    vi_tri_cuoi = 0
    for i in range(n):
        if f[i] == max_f:
            vi_tri_cuoi = i
            break
            
    ket_qua = []
    while vi_tri_cuoi != -1:
        ket_qua.append(mang[vi_tri_cuoi])
        vi_tri_cuoi = truoc[vi_tri_cuoi]
        
    return ket_qua[::-1]

mang_nhap = list(map(int, input().split()))
ket_qua = tim_day_con_tang_khong_lien_tuc_dai_nhat(mang_nhap)
print(*(ket_qua))
```
